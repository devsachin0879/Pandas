###### Dropping missing values using dropna()

```python
df = pd.DataFrame({
    'First Score': [100,90,np.nan,95],
    'Second Score': [30,np.nan,45,56],
    'Third Score': [52,40,80,98],
    'Fourth Score': [np.nan,np.nan,np.nan,65],
})
df
```

```python
df.dropna()
```

```python
df
```

```python
## drop a row whose all data is missing or contain null values(NaN)

df.dropna(how = 'all')
```

```python
## dropping COLUMNS with atleast 1 null value

df.dropna(axis = 1)
```

```python
# Dropping Rows with at least 1 null value in CSV file

df = pd.read_csv("employees.csv")
new_data = df.dropna(axis = 0, how = 'any')
new_data
```

```python
print("Old data frame length: ",len(df))
print("New data frame length: ",len(new_data))
print("Num of rows with atleast 1NA value: ",(len(df) - len(new_data)))
```

### pandas merge()

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35]
})
df1
```

```python
df2= pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "humidity": [65,68,75]
})
df2
```

```python
df3 = pd.merge(df1,df2,on = 'city')
df3               
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando","baltimore"],
    "temperature": [21,14,35,32]
})
df1
```

```python
df2= pd.DataFrame({
    "city": ["new york","chicago","san francisco"],
    "humidity": [65,68,71]
})
df2
```

```python
## intersection (equivalent to inner join)
df3 = pd.merge(df1,df2,on="city")
df3
```

```python
##outer
df3 = pd.merge(df1,df2,on="city",how= "outer")
df3
```

```python
### left 
df3 = pd.merge(df1,df2,on="city",how= "left")
df3
```

```python
##right
df3 = pd.merge(df1,df2,on="city",how= "right")
df3
```

```python
##outer
df3 = pd.merge(df1,df2,on="city",how= "outer",indicator = True)
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando","baltimore"],
    "temperature": [21,14,35,32],
    "humidity": [65,68,71,75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city")
df3
```

```python
df3 = pd.merge(df1,df2,on="city",suffixes = ('_left','_right'))
df3
```
